```python
import pandas as pd
#First, we load the dataset.
adult_census = pd.read_csv("..//M1.03/adult-census.csv")
```

```python
target_name = "class"
target = adult_census[target_name]
data = adult_census.drop(columns=[target_name, "education-num"])
```

```python
#we use this selector to get only the columns containing strings (column with object dtype) that correspond to categorical features in our dataset.
from sklearn.compose import make_column_selector as selector
categorical_columns_selector = selector(dtype_include=object)
categorical_columns = categorical_columns_selector(data)
data_categorical = data[categorical_columns]
```

```python
#Define a scikit-learn pipeline composed of an OrdinalEncoder and a LogisticRegression classifier.
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LogisticRegression

categorical_pipeline = make_pipeline(
    OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1),
    LogisticRegression(max_iter=500))
```

```python
import numpy as np
from sklearn.model_selection import cross_validate
# our model is now defined. Evaluate it using a cross-validation using
cv_results = cross_validate(
    categorical_pipeline,
    data_categorical,
    target,
    cv=5,
    error_score="raise",
    return_train_score=True
)
mean_acc = cv_results["test_score"].mean()
std_acc = cv_results["test_score"].std()
print(f"The mean cross-validation accuracy is: {mean_acc:.3f} ± {std_acc:.3f}")
```

```python
from sklearn.preprocessing import OneHotEncoder
model_ohe = make_pipeline(OneHotEncoder(handle_unknown="ignore"),LogisticRegression(max_iter=1000))
```

```python
cv_ordinal = cross_validate(
    categorical_pipeline,
    data_categorical,
    target,
    cv=5,
    error_score="raise"
)

cv_ohe = cross_validate(
    model_ohe,
    data_categorical,
    target,
    cv=5,
    error_score="raise"
)

mean_ord = cv_ordinal["test_score"].mean()
std_ord  = cv_ordinal["test_score"].std()

mean_ohe = cv_ohe["test_score"].mean()
std_ohe  = cv_ohe["test_score"].std()

print(f"OrdinalEncoder -> accuracy: {mean_ord:.3f} ± {std_ord:.3f}")
print(f"OneHotEncoder  -> accuracy: {mean_ohe:.3f} ± {std_ohe:.3f}")
```
